```python
#code to decode emojis to its real meaning

import unicodedata
from unidecode import unidecode
def deCodeEmoji(inputString):
    returnString = ""
    
    for character in inputString:
        try:
            character.encode("ascii")
            returnString += character
        except UnicodeEncodeError:
            temp = unidecode(str(character))
            if temp != '':
                returnString += temp
            else:
                try:
                     returnString += "[" + unicodedata.name(character) + "]" 
                except ValueError:
                     returnString += "[x]"

    return returnString
```

```python
translator = Translator()  #These two have been used for translation purpose
gs = goslate.Goslate()
```

```python
#This function is used to calculate the polarity and subjectivity of the text passed to it and stores its values into indpdict and indsdict respectively.

@jit      #jit has been used for the faster execution by utilizing the GPU
def Polarity_Subjectivity_fun(text, indpdict, indsdict, indhtag):
    text=deCodeEmoji(text)
    t1=translator.translate(text)
    #t1=gs.translate(text,'en')
    blob=TextBlob(str(text))
    for sentence in blob.sentences:
        p=sentence.sentiment.polarity
        s=sentence.sentiment.subjectivity
        if p==s and p==0:
             continue
        else:
            indpdict[indhtag].append(p)
            indsdict[indhtag].append(s)
```
